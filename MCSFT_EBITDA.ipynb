```python
data_ = data.diff(1)
data_ = data_.dropna()

plt.plot(data_)
plt.show()
```

```python
functs.lag_plot(data_, 5)
```

CHECKING FOR SEASONALITY
******

```python
from statsmodels.tsa.seasonal import STL
stl = STL(data_['EBITDA'], seasonal=13)
decomp = stl.fit()
fig = decomp.plot()
```

```python
years = ['2016', '2017', '2018', '2019', '2020', '2021']

seasons = []
s_1 = data_.loc['2016-01-01':'2017-01-01']
s_2 = data_.loc['2017-01-01':'2018-01-01']
s_3 = data_.loc['2018-01-01':'2019-01-01']
s_4 = data_.loc['2019-01-01':'2020-01-01']
s_5 = data_.loc['2020-01-01':'2021-01-01']
s_6 = data_.loc['2021-01-01':'2021-12-31']
s_7 = data_.loc['2022-01-01':]

seasons.append(s_1)
seasons.append(s_2)
seasons.append(s_3)
seasons.append(s_4)
seasons.append(s_5)
seasons.append(s_6)
seasons.append(s_7)

figure = plt.figure(figsize=(12, 15))
for season in seasons:
    plt.plot(season.index.month, season.EBITDA)
    plt.xlabel('Month')
    plt.ylabel('EBITDA')
plt.legend([year for year in years])
plt.show()
```

MODEL ( SARIMAX )
-----
*****

```python
import pmdarima
```

```python
auto = pmdarima.auto_arima(data, information_criterion="aicc", stationary=False, max_d=2, m=4, seasonal = True, maxd=2, trace=True)
print(auto.summary())
```

```python
plt.figure()
auto.plot_diagnostics(figsize=(20,10))
plt.show()
```

```python
resid = auto.resid()
resid = pow(resid, 2)
sm.graphics.tsa.plot_acf(resid)
plt.show()
sm.graphics.tsa.plot_pacf(resid, lags=10)
plt.show()

print(resid)
```

PREDICTIONS
****

```python
prediction_IS = auto.predict_in_sample(data)
prediction = prediction_IS
prediction.columns = ['EBITDA']
z = plt.plot(data, label = 'real')
x = plt.plot(prediction, label='prediction')
plt.show()
```

```python
prediction_OS = auto.predict(n_periods=4)
v = plt.plot(data, label='real')
b = plt.plot(prediction_OS, label='prediction')
plt.show()
```
